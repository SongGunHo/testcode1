```python
import numpy as np
fish_length = np.array([25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0])
fish_weight = np.array([242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9])
```

```python
fish_data = np.column_stack((fish_length, fish_weight))
fish_data
```

```python
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
fish_target
```

```python

from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    fish_data, fish_target, random_state=42, test_size=0.25, stratify=fish_target
)
# random_statre = - 랜덤 분할 결과를 항상 동일하게 유지하도록 고정
#stratify=fish_target = - 클래스 비율 유지. 즉, 도미와 빙어의 비율이 train/test에 고르게 분배되도록 해줌 
# test_size=0.25, = -  데이터의 25%를 테스트용, 나머지 75%를 학습용으로 나눔
```

```python
test_target
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
kn.fit(train_input, train_target) # 학습
```

```python
kn.score(test_input, test_target)
```

```python

kn.predict([[30, 890], [10, 12]]) # 정확도 
```

```python

kn.predict([[25, 150]]) # 도미 - 25cm, 150g
```

```python
import matplotlib.pyplot as plt
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(25,150, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
"""
- 무게이든 길이든 모두 중요한 특성!, 수치가 크다고 해서 특정 특성을 중요하게 볼 필요 X
- 수치를 동일한 기준의 범위로 변환 - 정규화 
    예) 표준점수
    분산 : (측정값 - 평균)^2  + (측정값 - 평균) + (측정값 - 평균) ... / 총 샘플 갯수 
    표준편차 : 분산에 제곱근
    표준점수 : (측정값 - 평균) +  (측정값 - 평균) + (측정값 - 평균) ... / 표준편차 
"""

# kn.kneighbors([[...], [...]]) : 반환값 (거리, 근처이 있는 요소 인덱스)
```

```python

distances, indexes = kn.kneighbors([[25, 150]])
```

```python
indexes
```

```python
import matplotlib.pyplot as plt
plt.scatter(train_input[:, 0], train_input[:, 1])
plt.scatter(25,150, marker='^')
plt.scatter(train_input[indexes, 0], train_input[indexes, 1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
mean = np.mean(train_input , axis=0) # 행 별 (특성별) 평균 
std  = np.std(train_input , axis=0) # 행별 (특성별 ) 표쥰 편자 
```

```python
train_scaled = (train_input - mean ) / std # 표준 점수
```

```python
kn.fit(train_scaled, train_target)
```

```python
fish = np.array(([25, 150]))
fish_scaled = (fish-mean) /std # 각 특성 값에서 학습테이터의 평군을 빼고 표준 편자 (std)  로 나누면 결과 값은 0, 표준 퍈자 1인 스케일로 변환 돈다 
fish_scaled # 정규화 된 생선 데이터 담는  번수 
```

```python
distances = indexes = kn.kneighbors(fish_scaled)
```

```python
kn.predict(fish_scaled)
```
